# Water Portability


### Water Portability:-
    Water portability refers to the quality and safety of water for drinking and everyday use. Safe drinking water, or potable water, should meet health standards, being free from harmful contaminants such as pathogens, chemicals, and physical impurities.

### Importance of Portability:-

### Problems Associated with non-potability of water:- There are number of problems associated with the non-potability of the water,so lets discuss them sectrowise

#  Health Sector:
1. **Waterborne Diseases:** Consumption of non-potable water can lead to diseases such as cholera, dysentery, typhoid, and giardia. These diseases are caused by pathogens like bacteria, viruses, and protozoa present in contaminated water.

1. **Toxic Substances:** Non-potable water may contain harmful chemicals such as lead, arsenic, and pesticides. Long-term exposure to these substances can cause serious health issues, including cancer, neurological disorders, and reproductive problems.

2. **Gastrointestinal Problems:** Drinking contaminated water can cause immediate effects such as nausea, vomiting, diarrhea, and stomach cramps, which can be particularly dangerous for vulnerable populations like children, the elderly, and those with weakened immune systems.

3. **Heavy Metals Poisoning:
Fluorosis:** Metals like lead and mercury can accumulate in the body over time, leading to severe health effects such as kidney damage, developmental delays in children, and cognitive impairments.

# Environment Sector:-
1. **Ecosystem Damage:** Contaminated water can harm aquatic ecosystems, killing fish and other aquatic life and disrupting the natural balance of ecosystems.

2. **Soil Contamination:** Using non-potable water for irrigation can lead to soil contamination, affecting crop quality and safety, and potentially entering the food chain.

3. **Biodiversity Loss:** Pollution from non-potable water sources can lead to a loss of biodiversity, affecting plants and animals that depend on clean water for survival.

# Socio-Economic Sector:-
1. **Economic Burden:** The costs associated with treating diseases caused by non-potable water can be significant, impacting individuals, families, and healthcare systems.

2. **Reduced Productivity:** Health problems from non-potable water can reduce productivity due to illness and absenteeism from work or school.

3. **Access Issues:** In regions with limited access to potable water, the time and effort required to obtain safe water can impact daily activities and economic opportunities.

# Infrastructure Sector
1. **Corrosion and Scaling:** Non-potable water with high levels of hardness or certain chemicals can cause corrosion and scaling in plumbing systems, leading to increased maintenance costs and potential infrastructure failures.

2. **Water Treatment Challenges:** High levels of contaminants in water sources can make it more difficult and expensive to treat the water to make it safe for consumption.

* ### In summary, the potability of water is integral to health, environmental sustainability, economic development and social well-being. Ensuring access to safe drinking water is a priority for improving global health and achieving sustainable development.

# Domain View:- 
> In this we will see every feature's informaton and realtion of it with potability of water

**1. PH:-**
> **pH** is a measure of the acidity or alkalinity of a substance, expressed on a scale ranging from 0 to 14.
> The pH of drinking water should ideally be between **6.5 and 8.5**,in our case potability of water is depends on PH

**2**. **Hardness**:- 
> **Water hardness** is a measure of the concentration of **calcium and magnesium ions** in water. It is an important water quality parameter that affects various aspects of water use in households, industries, and agriculture.For potable water the hardness should be in range of **0-120 ppm**


**3. Solids:-**
> Solids in water refer to the total amount of **suspended and dissolved substances** present in the water such as **minerals, salts, and metals**.The **WHO** suggests that water with Total dissolved solids levels up to **1000 mg/L** is generally acceptable to consumers, although it may have a noticeable taste.

**4. Chloramines:-**
> Chloramines are chemical compounds that contain **chlorine and ammonia**. The U.S. Environmental Protection Agency (EPA) sets the MRDL for chloramines in drinking water at **4.0 mg/L or ppm** . This level is considered safe for consumption and effective for disinfection purposes.

**5. Sulphate:-**
> **Sulfate** is a naturally occurring ion found in various mineral deposits and can be present in both surface and groundwater.The presence of sulfate in water can have several implications for water quality and human health.The WHO suggests that concentrations of sulfate below **250 mg/L** is safe to consume.

**6.Conductivity:-**
> Conductivity in water is a measure of its ability to conduct electrical current, which is directly related to the presence of dissolved ions such as **sulfate, chloride, nitrate, sodium, magnesium, calcium, and others**. The higher the concentration of these ions, the higher the conductivity of the water.Higher conductivity indicates higher levels of **dissolved salts and minerals**.For potable water, conductivity values typically range between **50 and 1500 µS/cm.**

**7.Organic_carbon:-**
> Organic carbon in water refers to carbon-containing compounds derived from living organisms or their remains. It can come from natural sources like decaying plants and animals, as well as anthropogenic sources like wastewater and agricultural runoff.The World Health Organization (WHO) suggests that the concentration of DOC in drinking water should be below **3 mg/L** to prevent the formation of undesirable tastes and odors and to minimize the potential for the formation of disinfection byproducts.

**8. Trihalomethanes:-**
> Trihalomethanes (THMs) are a group of chemical compounds that are formed as byproducts when chlorine or other disinfectants react with organic matter in water, particularly during the treatment of drinking water. THMs are considered disinfection byproducts (DBPs) and can have implications for water quality and health.The U.S. Environmental Protection Agency (EPA) has set a maximum contaminant level (MCL) for total trihalomethanes (TTHMs) at **80 micrograms per liter (µg/L) or 0.080 milligrams per liter (mg/L)** in treated drinking water


**9. Turbidity:-**
> Turbidity is a measure of the cloudiness or haziness of a liquid caused by suspended particles, which are generally invisible to the naked eye. In water, turbidity is often caused by particles such as **silt, clay, organic matter, algae, and other microscopic organisms**. Turbidity can affect water quality and is an important parameter in assessing the suitability of water for various uses, including drinking water.The **World Health Organization (WHO)** recommends that turbidity in drinking water **should not exceed 5 NTU at any time and should ideally be below 1 NTU** for aesthetic reasons.

### Outlier Treatment

```python
## Defining outlier function
def outliers(df,column):
    Q1=df[column].quantile(0.25)
    Q3=df[column].quantile(0.75)
    IQR=Q3-Q1
    Lower_Limit=Q1-1.5*IQR
    Upper_Limit=Q3+1.5*IQR
    return Lower_Limit,Upper_Limit
```

```python
Column_Names=[]
Outlier_Per=[]
for column in Master_Dataset.columns:
    ## Getting Upper and lower limit for each columns
    Lower_limit,Upper_limit=outliers(Master_Dataset,column)
    ## Checking the outliers size
    Outliers_data=Master_Dataset[(Master_Dataset[column]<Lower_limit)|(Master_Dataset[column]>Upper_limit)].shape[0]
    Outliers_percentage=(Outliers_data)
    Outliers_percentage=round(Outliers_percentage,2)
    Outliers_percentage
    Column_Names.append(column)
    Outlier_Per.append(Outliers_percentage)
outliers_size=pd.DataFrame({"Column Name":Column_Names,"Outlier Numbers":Outlier_Per})
outliers_size
outliers_size['outliers_percentage']=(outliers_size['Outlier Numbers']/len(Master_Dataset))*100
outliers_size=outliers_size.sort_values(by='outliers_percentage',ascending=False)
outliers_size
```

### So as we can seen that there are not to much outliers in the data as compared to the datasize so we can drop it 

### Dropping outliers

```python
def outliers_removal(Dataset):
    Master_Dataset_New=[]
    Updated_dataset=Dataset
    for col in Dataset.columns:
        Lower_limit,Upper_limit=outliers(Updated_dataset,col)
        Updated_dataset=Updated_dataset[(Updated_dataset[col]>Lower_limit)&(Updated_dataset[col]<Upper_limit)]
        Master_Dataset_New.append(Updated_dataset)
    return Master_Dataset_New
```

```python
mtd=outliers_removal(Master_Dataset)
Master_Dataset_New=mtd[-1]
print(f'Total size of data after outliers removal is {Master_Dataset_New.shape}')
```

## Modelling

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,roc_curve,roc_auc_score,auc
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
#from sklearn.svm import svc
import xgboost as xgb
import ast
```

```python
def model_evaluaion(X_train,X_test,Y_train,Y_test):
    Models_names=['Logistic regression','Descion Tree','Random Forest',
                  'Ada boost']
    ## Creating model objects
    le=LogisticRegression()
    dt=DecisionTreeClassifier()
    rf=RandomForestClassifier()
    #svm=svc()
    adbst=AdaBoostClassifier()
    #xg=xgb()
    ## Basic Models
    model_set1=[le,dt,rf,adbst]
    Model_Accuracy=[]
    Model_precesion=[]
    Model_recall=[]
    Model_F1_Score=[]
    cm=[]
    predicted_value=[]
    for mod in model_set1:
        mod.fit(x_train,y_train)
        result=mod.predict(x_test)
        accuracy=accuracy_score(y_test,result).round(2)
        precision=precision_score(y_test,result).round(2)
        recall=recall_score(y_test,result).round(2)
        F1_Score=f1_score(y_test,result).round(2)  
        confusion_mat=confusion_matrix(y_test,result)
        fpr, tpr, _ = roc_curve(y_test,result)
        auc =roc_auc_score(y_test,result)
        Model_Accuracy.append(accuracy)
        Model_precesion.append(precision)
        Model_recall.append(recall)
        Model_F1_Score.append(F1_Score)
        predicted_value.append(result)
        cm.append(confusion_mat)
    Result_Table=pd.DataFrame({"Model Name":Models_names,"Accuracy":Model_Accuracy,
                               "Precision":Model_precesion,"Recall":Model_recall,
                               "F1 Score":Model_F1_Score,"Predictions":predicted_value,"cm":cm})
    Result_Table.sort_values(by='Accuracy',ascending=False,inplace=True)
    return Result_Table
```

```python
Result_Table=model_evaluaion(x_train,x_test,y_train,y_test)
num_spaces = 10
result=Result_Table.iloc[:,0:6]
print("Model Performance\n")
print(result)
print("\n")
print('Heat Map\n')

array_list=[]
## Creating Confusion_Matrix
for j in Result_Table['cm']:
    array_list.append(j)
array_list
## Model List
models=[i for i in Result_Table['Model Name']]
    
# Create a 2x2 subplot grid
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
for i,(con,mn) in enumerate(zip(array_list,models)):
    row = i // 2
    col = i % 2
    #con=np.array(ast.literal_eval(con))
    sns.heatmap(con, annot=True, ax=axs[row, col],fmt='g')
    axs[row, col].set_title(mn)
    
# Adjust the layout
plt.tight_layout()

# Display the charts
plt.show()

## printing ROC AUC Curve
print("ROC AUC_Curve\n")

for i,(cn,dte) in enumerate(zip(Result_Table['Model Name'],Result_Table['Predictions'])):
    row = i // 2
    col = i % 2
    ## Plotting roc curve
    fpr, tpr, _ =roc_curve(y_test,dte)
    plt.plot(fpr, tpr, label=f'{cn} (AUC = {auc(fpr, tpr):.2f})')
     
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

### Still our accuracy not increases,now checking with cross validation 

```python
## Importung Laibaries
from sklearn.model_selection import KFold,cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
## Used Models
def cross_models(X,Y,N):
    lr=LogisticRegression()
    DT=DecisionTreeClassifier()
    RF=RandomForestClassifier()
    knn=KNeighborsClassifier(n_neighbors=5)
    models_list=[lr,DT,RF,knn]
    Model_Name=[]
    Scores=[]
    #K-fold setup
    kf=KFold(n_splits=N,shuffle=True,random_state=42)
    for model in models_list:
        # Perform cross-validation
        score = cross_val_score(model, X, Y, cv=kf)
        score=np.mean(score).round(2)
        Scores.append(score)
        Model_Name.append(model)
    
    ## Creating  DataFrame
    Result_Table=pd.DataFrame({"Model Name":Model_Name,"Model Score":Scores})
    Result_Table.sort_values(by='Model Score',inplace=True,ascending=False)
    return Result_Table
```

```python
cross_models(X,Y,5)
```
